```python
import pandas as pd
```

```python
rating_df = pd.read_csv('drive/MyDrive/1004/ratings_large.csv')
# train_small = pd.read_csv("/scratch/ll4764/train_small.txt", sep=" ", header=None, names=["user_id", "movie_id", "rating", "timestamp"])
train_test = pd.read_csv("drive/MyDrive/1004/train_test.txt", sep=" ", header=None, names=["user_id", "movie_id", "rating", "timestamp"])
train_val = pd.read_csv("drive/MyDrive/1004/train_val.txt", sep=" ", header=None, names=["user_id", "movie_id", "rating", "timestamp"])
```

```python
small_ratings_2 = rating_df.groupby('userId').filter(lambda x: len(x) > 10)
rating = small_ratings_2.groupby('movieId').filter(lambda x: len(x) > 10)
```

```python
userid = rating['userId'].unique()
train_userid = userid[: int(0.6 * len(userid))]
```

```python
train = rating[rating['userId'].isin(train_userid)]
frames = [train, train_val, train_test]
result = pd.concat(frames)
train.to_csv('train_full.txt', index=False, header=False, sep=' ')
train
```
