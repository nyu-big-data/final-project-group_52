```python
import pandas as pd
import operator
import numpy as np
```

```python
train_small = pd.read_csv("drive/MyDrive/1004/train_small.txt", sep=" ", header=None, names=["user_id", "movie_id", "rating", "timestamp"])
test_small = pd.read_csv("drive/MyDrive/1004/test_small.txt", sep=" ", header=None, names=["user_id", "movie_id", "rating", "timestamp"])
val_small = pd.read_csv("drive/MyDrive/1004/val_small.txt", sep=" ", header=None, names=["user_id", "movie_id", "rating", "timestamp"])
```

```python
train_rating_sum = train_small['rating'].sum()
avg_rating = train_rating_sum / len(train_small)
avg_rating
```

```python
sorted_train = dict(sorted(train_small['rating'].items(), key=operator.itemgetter(1), reverse=True))
movie_100 = list(sorted_train.keys())[: 100]
```

```python
total_val_num = len(val_small)
total_test_num = len(test_small)
```

```python
def MAP_rmse(df):
  map_cnt = 0
  mse_cnt = 0
  mse = 0
  P = 0
  for i in df['user_id'].unique():
    table = df[df['user_id']==i]
    temp = table.movie_id.to_list()
    
    for j in temp:
      if j not in movie_100:
        map_cnt += 1
      else:
        rating = table[table['movie_id']==j].rating.to_list()[0]
        mse = mse + (rating - sorted_train.get(j)) ** 2
        if rating >= avg_rating:
          P += 1

        mse_cnt += 1
        map_cnt += 1
            
  print("MAP score: ", P/map_cnt, "\nRMSE score: ", (mse/mse_cnt) ** 0.5)
    
```
